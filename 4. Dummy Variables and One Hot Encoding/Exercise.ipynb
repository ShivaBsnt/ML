```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datasets/carprices.csv")
df
```

```python
# LABEL ENCODE CAR MODEL
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df_le = df
```

```python
df_le
```

```python
df_le['Car Model'] = le.fit_transform(df_le['Car Model'])
```

```python
df_le
```

```python
X = df_le[['Car Model', 'Mileage', 'Age(yrs)']]
```

```python
X
```

```python
y = df_le[['Sell Price($)']]
y
```

```python
plt.scatter(X,y)
```

```python
# Transform Column into Dummy Variables
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(
    transformers=[
        ('car_model_enc', OneHotEncoder(drop="first", handle_unknown="ignore"), ['Car Model']),  
    ],
    remainder = "passthrough"
)
```

```python
X = ct.fit_transform(X)
```

```python
X
```

```python
reg = linear_model.LinearRegression()
reg.fit(X,y.values)
```

```python
reg.predict([[0,1,45000,4]])
```

```python
reg.predict([[1,0,86000,7]])
```

```python
reg.score(X,y)
```
