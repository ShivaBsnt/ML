```python
import pandas as pd
```

```python
df = pd.read_csv("datasets/titanic.csv")
df.head()
```

```python
df.columns
```

```python
selected_columns = ['Pclass', 'Sex', 'Age', 'Fare', 'Survived']
df = df[selected_columns]
df.head()
```

```python
# CHECK FOR NULL VALUES
df.info()
```

```python
# GET MEAN FOR AVERAGE
mean_age = df['Age'].mean()
mean_age
```

```python
# FILL NULL VALUE OF AGE
df['Age']=df['Age'].fillna(mean_age)
df.head()
```

```python
df.info()
```

```python
# LABEL ENCODE SEX COLUMN AS IT IS NOMINAL
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Sex']=le.fit_transform(df['Sex'])
df.head()
```

```python
# INDEPENDENT VARIABLE
X=df[['Pclass', 'Sex', 'Age', 'Fare']]
X.head()
```

```python
# DEPENDENT VARIABLE
y = df[['Survived']]
y.head()
```

```python
# SINCE THE GENDER COLUMN IS IN TEXT, DO ONE-HOT ENCODING
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(
    transformers=[
        ('gender_enc', OneHotEncoder(drop='first', handle_unknown='ignore'), ['Sex'])
    ],
    remainder = 'passthrough'
)
```

```python
X=ct.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_tesy = train_test_split(X, y)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_tesy)
```
