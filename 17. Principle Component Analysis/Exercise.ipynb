```python
df5 = pd.get_dummies(df, drop_first=True)
df5.head()
```

```python
X = df5.drop("HeartDisease",axis='columns')
y = df5.HeartDisease

X.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=30)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

model_params = {
    'logistic_regression': {
        'model': LogisticRegression(),
        'params': {
            'solver': ['lbfgs', 'liblinear'],
            'C': [1,3,6,9,12]
        }
    },
    'decision_tree':{
        'model': DecisionTreeClassifier(),
        'params': {
            'criterion': ['gini', 'entropy', 'log_loss']
        }
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators': [5,10,15,20,25]
        }
    },
    'svm': {
        'model': SVC(),
        'params': {
            'C': [1, 10, 20, 30, 40],
            'kernel': ['rbf', 'linear'],
            'gamma':['scale', 'auto']
        }
    }
}
```

```python
from sklearn.model_selection import GridSearchCV
scores = []
for model_name, model_param in model_params.items():
    search = GridSearchCV(model_param['model'],model_param['params'],cv=5,return_train_score=False)
    search.fit(X_train, y_train)
    scores.append({
        'model': model_name,
        'best_score': search.best_score_,
        'best_param': search.best_params_
    })
scores
```

```python
# Use PCA to reduce dimension
from sklearn.decomposition import PCA
pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

```python
from sklearn.svm import SVC
model_svm = SVC(C=1, gamma='scale', kernel='rbf')
model_svm.fit(X_train_pca, y_train)
model_svm.score(X_train_pca, y_test)
```
