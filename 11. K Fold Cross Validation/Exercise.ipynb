```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
# loading iris dataset to iris variable
iris = load_iris()
iris.data[:5]
```

```python
# IMPORT MODELS TO TRAIN
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
# TEST SCORE FOR DIFFERENT MODEL
# Logistic Regression
from sklearn.model_selection import cross_val_score
np.average(cross_val_score(LogisticRegression(max_iter=500), iris.data, iris.target, cv=3))
```

```python
# Support Vector Machine
np.average(cross_val_score(SVC(), iris.data, iris.target, cv=3))
```

```python
# Random Forest
np.average(cross_val_score(RandomForestClassifier(), iris.data, iris.target, cv=3))
```
