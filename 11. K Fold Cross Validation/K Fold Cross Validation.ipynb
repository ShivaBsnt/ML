```python
import pandas as pd
from sklearn.datasets import load_digits
```

```python
digit = load_digits()
```

```python
# SPLIT DATA FOR TRAIN AND TEST
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digit.data, digit.target, test_size=0.3)
```

```python
# LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
# SUPPORT VECTOR MACHINE
from sklearn.svm import SVC
svm = SVC()
svm.fit(X_train, y_train)
svm.score(X_test, y_test)
```

```python
# RANDOM FOREST
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
# K-FOLD
from sklearn.model_selection import KFold
kf = KFold(n_splits=3)
kf.get_n_splits()
```

```python
# SPLIT DATA INTO K FOLD
for i, (train_index, test_index) in enumerate(kf.split([1,2,3,4,5,6,7,8,9])):
    print(f"Index: {i}, Train Index: {train_index}, Test Index: {test_index}")
```

```python
# RETURNS SCORE OF THE MODEL
def get_score(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    return model.score(X_test, y_test)
```

```python
get_score(RandomForestClassifier(), X_train, X_test, y_train, y_test)
```

```python
# USING STRATIFIELDKFOLD TO ENSURE THAT EACH FOLD MAINTAINS THE SAME CLASS DISTRIBUTION
from sklearn.model_selection import StratifiedKFold
s_fold = StratifiedKFold(n_splits=3)
```

```python
lr_score = []
svm_score = []
rf_score = []

for train_index, test_index in s_fold.split(digit.data, digit.target):
    X_train,X_test,y_train,y_test = digit.data[train_index],digit.data[test_index], digit.target[train_index], digit.target[test_index]
    lr_score.append(get_score(LogisticRegression(), X_train, X_test, y_train, y_test))
    svm_score.append(get_score(SVC(), X_train, X_test, y_train, y_test))
    rf_score.append(get_score(RandomForestClassifier(), X_train, X_test, y_train, y_test))
```

```python
# VALIDATION USING CROSS VAL
from sklearn.model_selection import cross_val_score
cross_val_score(LogisticRegression(max_iter=5000), digit.data, digit.target, cv=3)
```

```python
cross_val_score(SVC(), digit.data, digit.target, cv=3)
```

```python
cross_val_score(RandomForestClassifier(), digit.data, digit.target, cv=3)
```
