### BIAS
Bias is an error due to overly simplistic assumptions in a machine learning model.

### High Bias
It is a condition where the model underfits the data, and does not understand the true relationship.

### Low Bias
It is a condition where the model is sufficiently complex and captures the relationship between the data.

### Variance
Variance is an error due to a small fluctuation in the training dataset.

### High Variance
It is a condition where the model overfits the data, failing to understand the relationship and instead memorizing the data.

### Low Variance 
It is a condition where the model is general enough to understand the pattern and relationship in the data.
