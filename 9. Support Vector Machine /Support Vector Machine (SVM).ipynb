```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.data[0]
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
iris.target_names
```

```python
df['target'] = iris.target
```

```python
df
```

```python
#setosa
df[df['target']==0]
```

```python
df['target_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
# KEEP THE DIFFERENT FLOWER IN DIFFERENT DATAFRAME
df0 = df[df['target']==0]
df1 = df[df['target']==1]
df2 = df[df['target']==2]
```

```python
# plot the different flowers in a single graph
import matplotlib.pyplot as plt
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color="red", marker="+")
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color="blue", marker="*")
# plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color="black", marker=".")
```

```python
# plot the different flowers in a single graph
import matplotlib.pyplot as plt
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color="red", marker="+")
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color="blue", marker="*")
# plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'], color="black", marker=".")
```

```python
X = df[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']]
y = df['target']
```

```python
X.head()
```

```python
y.head()
```

```python
# SPLIT DATA FOR TRAIN AND TEST
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(X,y)
```

```python
# Train the data in model
from sklearn.svm import SVC
model = SVC()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
