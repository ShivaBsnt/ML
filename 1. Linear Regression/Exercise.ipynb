```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("datasets/canada_per_capita_income.csv").rename(columns={
    'per capita income (US$)': 'per_capita_income'
})
```

```python
plt.xlabel("year")
plt.ylabel("per capita income US$")
plt.title("year vs per capita income")
plt.scatter(df.year, df.per_capita_income)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df.per_capita_income)
```

```python
reg.predict(pd.DataFrame({'year': ['2020']}))
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("per capita income US($)")
plt.title("Per capita income Vs Year")
plt.scatter(df.year, df.per_capita_income)
plt.plot(df.year, reg.predict(df[['year']]), color="blue")
```
